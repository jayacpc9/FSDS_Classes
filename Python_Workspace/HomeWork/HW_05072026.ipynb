# Binary Numbers
### See example : to convert value dynamically
### int(Value, base)

```python
binVal = bin(30)
print(binVal)
originalVal = int(binVal,2)
print(originalVal)
```

# Octal Numbers

```python
num = 25
octVal=oct(num)
print(octVal)
originalVal = int(octVal,8)
print(originalVal)
```

# HexaDeecimal Numbers

```python
num = 25
hexVal = hex(num)
print(hexVal)

originalVal = int(hexVal, 16)
print(originalVal)
```

# Swaping 2 Variables 
### For example :  if  (a,b) = (5,6)  swap them as (a,b) = (6,5)

```python
a = 5
b = 6
print(f"*"*10+" Before")
print(f"a = {a}")
print(f"b = {b}")

# use a temp variable 
temp = a
a =b
b = temp

print(f"*"*10+" After")
print(f"a = {a}")
print(f"b = {b}")
```

# Swapping without using a 3rd variable

```python
a = 5
b = 6
print(f"*"*10+" Before")
print(f"a = {a}")
print(f"b = {b}")

# use a temp variable 

a = a +b
b = a - b
a =  a - b

print(f"*"*10+" After")
print(f"a = {a}")
print(f"b = {b}")
```

# Swapping 2 numbers using XOR 

```python
a = 51232524525
b = 6986987692345

print(f"*"*10+" Before")
print(f"a = {a}")
print(f"b = {b}")

binA = bin(a)
binB = bin(b)
print(f"a = {binA}")
print(f"b = {binB}")


a = a ^ b
print(f" xor ( a = a ^ b ) = {a}")

b = a ^ b
print(f" xor ( b = a ^ b ) = {b}")

a = a ^ b
print(f" xor ( a = a ^ b ) = {a}")

# use a temp variable 
print(f"*"*10+" After")
print(f"a = {a}")
print(f"b = {b}")
```
