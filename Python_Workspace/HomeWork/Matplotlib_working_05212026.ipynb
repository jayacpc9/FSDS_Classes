# Basic line plot
### Line plots  are the most common type of plot.
### They are great for showing trends over time or relationship between variables

```python
# create some sample data
x = np.linspace(0,10,100) # creates 100 points from 0 to 10 (linear data)
y = np.sin(x)

# smiple plot
plt.figure(figsize= (8,5))
plt.plot(x,y)
plt.xlabel('X values')
plt.ylabel("sin(X)")
plt.title("Sine Wave")
plt.grid(True)
plt.show()
```

```python
# multiple lines on the same plot
x = np.linspace(0,10,100) # creates 100 points from 0 to 10 (linear data)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.sin(x) * np.cos(x)

# smiple plot
plt.figure(figsize= (8,5))
plt.plot(x,y1,label ='sin(x)')
plt.plot(x,y2, label = 'cos(x)')
plt.plot(x,y3, label = 'sin(x) * cos(x)')
plt.xlabel('X values')
plt.ylabel("Y values")
plt.title("Multiple Functions")
plt.legend()
plt.grid(True, alpha = 0.5)
plt.show()
```

# customizing Plots: Colors, Styles and Markers
### matplotlib offers extensive customization options to make your plots visually appealing and informative.

```python
x = np.linspace(0,10,20)

plt.figure(figsize=(12,6))
plt.plot(x,x,'b-',label='Solid Line', linewidth=2) # b- = blue color with solid line
plt.plot(x,x+1,'r--',label='Dashed Line', linewidth=2) # r-- = Red color with Dashed line
plt.plot(x,x+2,'g-',label='Dash-dot Line', linewidth=2) # g- = Green color solid line
plt.plot(x,x+3,'m:',label='Dotted Line', linewidth=2) # m: = Magenta color with Dotted line

plt.plot(x,x+4,'co-',label='Circles', markersize=6) # co- =Cyan color with Circle markder on Solid line
plt.plot(x,x+5,'rs-',label='Squares', markersize = 6) # rs- = Red Square on solid line
plt.plot(x,x+6,'^-',label='Triangles', markersize = 6) # ^- = Triangle on solid line picks a random color
plt.plot(x,x+7,'*-.',label='* Dash-dot', markersize = 6) # *-. = * as marker on a Dash-Dot line
plt.plot(x,x+8,'x-',label='x Solid', markersize = 6) # x- = X on a Solid line

plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("Line styles and Markers")

plt.legend()
plt.grid(True,alpha = 0.5)
plt.show()

# Common format string : '[color][marker][line]'
#Colors : b(Blue),r(Red),g(Green),c(Cyan),m(Magenta), y(Yellow),k(Black)
# Markers : o(Circle), s(Square), ^ (Triangle), * (Star),+ (Plus), x(X)
# Lines : - (Solid),--(Dashed), -.(Dash-dot), : (Dotted)
```

```python
# using color names and hex codes
x = np.linspace(0,10,100)

plt.figure(figsize= (12,6))
plt.plot(x, np.sin(x), color= 'steelblue', linewidth = 2, label = 'SteelBlue')
plt.plot(x, np.cos(x), color ='#FF6B6B', linewidth = 2, label ='#FF6B6B (hex)')
plt.plot(x, np.sin(x*2), color= 'darkgreen', linestyle='--', linewidth = 2, label ='darkgreen')

plt.xlabel('X value', fontsize =12)
plt.ylabel('Y value', fontsize = 12)
plt.title("Custom Colors and Styles", fontsize = 14, fontweight= 'bold')
plt.legend(fontsize = 10)
plt.grid(True, alpha = 0.5)
plt.show()
```

```python
# scattered plot graph
np.random.seed(42)
x =np.random.randn(100)

y = (2 * x) + (np.random.randn(100) * 0.5)

plt.figure(figsize=(8,6))
plt.scatter(x, y, alpha = 0.6) # alpha controls the transparency
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title("Scatter Plot", fontsize = 12)

plt.grid(True, alpha=0.3)
plt.show()
```

```python
# scatter plot with different colors and sizes
np.random.seed(42)
x = np.random.randn(100)
y = np.random.randn(100)
colors = np.random.randn(100) # random color codes
sizes = 1000 * np.random.randn(100) # Random sizes

plt.figure(figsize= (10,6))
plt.scatter(x, y, c = colors,s = sizes,alpha=0.6,cmap= 'Set1')
plt.colorbar(label = 'Color Intensity')
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("Scatter Plot with color and size variation")
plt.grid(True,alpha= 0.3)
plt.show()
```

# Bar charts
### Bar charts are ideal for comparing categorical data or showing counts

```python
# Smimple bar Chart
categories = ['A','B','C','D','E']
values = [23,45,56,78,32]
plt.figure(figsize= (10,8))
plt.bar(categories,values, color= 'steelblue',edgecolor = 'black', linewidth = 1.2)
plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Simple Bar Chart")
plt.grid(True, alpha=0.3, axis='y')
plt.show()
```

```python
# help(plt.grid)
# Horizontal bar Chart
categories = ['A','B','C','D','E']
values = [23,45,56,78,32]
plt.figure(figsize= (10,8))
plt.barh(categories,values, color= 'coral',edgecolor = 'black', linewidth = 1.2)

plt.xlabel("Values")
plt.ylabel("Categories")
plt.title("Horizontal Bar Chart")
plt.grid(True, alpha=0.3, axis='x')
plt.show()
```

```python
# grouped bar chart
categories = ['Q1','Q2','Q3','Q4']
product_a = [20,35,30,35]
product_b = [25,32,34,20]
product_c = [30,25,30,25]

x = np.arange(len(categories))

plt.figure(figsize=(10,8) )
width = 0.25 # width of the bars in chart

plt.bar(x-width,product_a, width,label = 'Product A',color = '#FF6B6B')
#plt.bar(x, height, width,...) 
# parameters:
# x - width : x-position(Shift left by one bar width to position first group)
# product_a : height values for product A bars
# width : width of each bar (0.25)
# label : text for label
# color : bar fill color (hex code)

plt.bar(x,product_b, width, label="Product B", color ='#4ECDC4')
plt.bar(x+width,product_c,width, label="Product C", color = '#95E1D3')

plt.xlabel("Quarter")
plt.ylabel("Sales")
plt.title("Quaterly Sales Comparision")
plt.xticks(x,categories)
plt.legend()
plt.grid(True, alpha= 0.3, axis='y')
plt.show()
```

# Histograms
### Histograms show the distribution of numerical data by grouping values into bins.

```python
# Generate sample data
np.random.seed(42)
data = np.random.normal(100, 16,1000) # mean = 100, std = 15 and return 1000 points

# Basic Histogram
plt.figure(figsize=(10,8))
plt.hist(data, bins=30, color='skyblue',edgecolor='black', alpha = 0.6)
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.title("Histogram of Normal Distribution")
plt.legend()
plt.grid(True, alpha= 0.3, axis='y')
plt.show()
```

```python
# multiple histograms on the same plot

np.random.seed(42)
data1 = np.random.normal(100,15,1000)
data2 = np.random.normal(120,20,1000)

plt.figure(figsize=(10,6))
plt.hist(data1, bins=30, alpha=0.5,label='Data Set 1', color='blue')
plt.hist(data2, bins=30, alpha =0.5,label='Data Set 2', color='red')

plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Overlappnig Histogram")
plt.legend()
plt.grid(True, alpha=0.3, axis ='y')
plt.show()

```

# Subplot grid

```python
import numpy as np
# Create 2X2 subplot grid
x = np.linspace(0,10,100)
fig, axes = plt.subplots(2,2,figsize = (6,5))
fig.suptitle("Multiple Subplots", fontsize = 16 ,fontweight = 'bold')

print(type(x))
#Top Left : Line Plot
axes[0,0].plot(x, np.sin(x),'b-',linewidth=2)
axes[0,0].set_title('Sine Wave')
axes[0,0].grid(True, alpha=0.3)

#Top Right : Cosine
axes[0,1].plot(x, np.cos(x),'r-',linewidth=2)
axes[0,1].set_title('Cosine Wave')
axes[0,1].grid(True, alpha=0.3)


# Bottom Left : Scatter
np.random.seed(42)
x_scatter = np.random.randn(100)
y_scatter = np.random.randn(100)

axes[1,0].scatter(x_scatter,y_scatter,alpha=0.6, c=x_scatter, cmap ='viridis')
axes[1,0].set_title('Scatter Plot')
axes[1,0].grid(True, alpha=0.3)

# Bottom Right : Bar chart

categories = ['A','B','C','D']
values = [23,45,56,78]

axes[1,1].bar(categories, values, color = 'coral')
axes[1,1].set_title('Bar Chart')
axes[1,1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```
