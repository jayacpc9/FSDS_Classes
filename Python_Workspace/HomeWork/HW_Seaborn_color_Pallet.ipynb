# Seaborn color pallet

```python
import seaborn as sns
import matplotlib.pyplot as plt

# List of standard Seaborn palette names
deep_palettes = ["deep", "muted", "pastel", "bright", "dark", "colorblind"]

# Plot them to see what they look like
for pal in deep_palettes:
    print(f"Palette: {pal}")
    sns.palplot(sns.color_palette(pal))
    plt.show()
```

```python
from matplotlib import colormaps

# Get a list of all registered palette names
all_palettes = list(colormaps)
print(all_palettes)

for pal in all_palettes:
    if "jet" in pal.lower():
        continue
    
    print(f"Palette: {pal}")
    sns.palplot(sns.color_palette(pal))
    plt.show()
```

```python
sns.color_palette("Set2")
```

```python
sns.color_palette("rocket", as_cmap=True)
```

```python
uniform_pallet =["rocket", "mako", "flare", "crest"]
for pal in uniform_pallet:
    print(f"Palette: {pal}")
    cmap = sns.color_palette(pal, as_cmap=True)    
    display(cmap)
```
