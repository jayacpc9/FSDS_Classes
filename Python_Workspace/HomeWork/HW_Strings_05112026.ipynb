# join() and split()
## join()
### The `join()`method takes all the items in an iterable, like alist,dictionary,tuple or set,and joins them into a string. 
### You can also specify a separator.

```python
''.join(['My','Name','is','jaya'])
```

```python
'*'.join(['My','Name','is','jaya'])
```

```python
' '.join(['My','Name','is','jaya'])
```

```python
', '.join(['cats','rats','bats'])
```

# Split()
### The `split()`method splits a `string`into a `list`. 
### By default, it will use whitespace toseparate the items, but you can also set another character of choice:

```python
'My Name is jaya'.split()
```

```python
'My*Name*is*jaya'.split('*')
```

# The Count Method 
### Counts the number of occurrences of a given character or substring in the string it isapplied to. 
### Can be optionally provided start and end index.

```python
sentence = 'one sheep two sheep three sheep four'
sentence.count('sheep')
```

```python
sentence.count('e')
```

```python
sentence . count ( 'e' , 6 ) # returns count of e after 'one sh' i.e 6 chars since beginning of string  
```

```python
sentence . count ( 'e' , 7 )
```
