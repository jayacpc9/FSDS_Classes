# Univariate analysis 

```python
cvd_df['target'].nunique() # Return number of unique elements in the object.
```

```python
cvd_df['target'].unique()
# Return unique values of Series object.

# Uniques are returned in order of appearance. Hash table-based unique, therefore does NOT sort.

# Returns
# ndarray or ExtensionArray
# The unique values returned as a NumPy array. See Notes.

```

```python
cvd_df['target'].value_counts()
# Return a Series containing counts of unique values.

# The resulting object will be in descending order so that the first element is the most frequently-occurring element. Excludes NA values by default.
```

#### Comment on the output:
#### 1 = presence of Heart disease, there are 165 people with heart disease.
#### 0 = absence of Heat diseas, there are 138 people without any heart disease
 

```python
sns.set(style="whitegrid")
# plt.subplots(figsize=(8,6))
sns.countplot(data=cvd_df, x='target')
```

```python
cvd_df.groupby('sex')['target'].value_counts()
```

#### Analysis of output
#### Heart disease grouped by sex.
#### 0: Femal : 72 with heart disease and 24 absence of heart disease
#### 1: Male : 93 with heart disease and 114 absence of heart disease
