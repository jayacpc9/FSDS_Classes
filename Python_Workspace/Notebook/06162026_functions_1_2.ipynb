# 1. positional Arguments

```python
def person(name, age):
    print("name = ",name)
    print("age = ",age)

person("joe",45)
print("-"*20)
person(45,"joe")
```

# 2.Keyword arguments

```python
def person(name, age):
    print("name = ",name)
    print("age = ",age)

person(age=45,name="joe")
```

# you can use either a position argument or keyword argument method,

```python
def person(name, age,mobile):
    print("name = ",name)
    print("age = ",age)

person(age=45,name="joe",mobile=1234)
```

# KWARGS (keyworded Variable Length Arguments)

```python
def person(name, *data): # this is an example for args
    print(name)
    print(data)
   

person('Alex',36,'John',987767)

```

```python
def person(name, **data): # 2 ** is for kwargs, we can mix and use the keyword and variable length arguments
    print(name)
    print(data)
   

person('Alex',age=36,home_place='southcity',mob = 987767)
```

```python
def multiplication(**kwargs):
    print(kwargs)
    multiple = kwargs['multiple'] 
    limit = kwargs['limit']
    
    print(f"Multiple: {multiple}, Limit: {limit}")
    
    for i in range(1,limit):
        print(f"{multiple} * {i} = {(multiple*i)}")

multiplication(multiple=2,limit=10)
```

```python
def multiplication(**kwargs):
    # Loop through every key-value pair individually
    for key, value in kwargs.items():
        print(f"Passed Argument -> Name: {key}, Value: {value}")

multiplication(multiple=2, limit=10)
```
