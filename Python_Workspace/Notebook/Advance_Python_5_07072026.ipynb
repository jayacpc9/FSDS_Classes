```python
# 1. The variables exist in the same file
gravity = 9.8
speed = 100

def handle_user_input():
    # 2. The user types the setting name as a string (e.g., "gravity")
    user_string = input("What setting do you want to change? ")
    new_value = float(input("Enter new value: "))
    
    # 3. We use globals() because everything is in ONE file
    if user_string in globals():
        globals()[user_string] = new_value
        print(f"Updated {user_string}!")
    else:
        print("Setting not found.")

def main_game_loop():
    print(f"Current gravity is {gravity}")

# Running the game
handle_user_input()
main_game_loop()
```
