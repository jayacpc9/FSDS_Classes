```python
x = 4
r = x%2
if r==0:
    print("Even number")
```

```python
x = 4
r = x%2
if r==0:
    print("Even number")

if r==1:
    print("Odd Number")
```

```python
x = 5
r = x%2
if r==0:
    print("Even number")
else:
    print("Odd Number")
```
