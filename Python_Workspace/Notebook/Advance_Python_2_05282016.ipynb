```python
x = 4
r = x%2
if r==0:
    print("Even number")
```

```python
x = 5
r = x%2
if r==0:
    print("Even number")

# the numbe is odd and the condition is not handled
```

```python
x = 4
r = x%2
if r==0:
    print("Even number")

if r==1:
    print("Odd Number")
```

```python
x = 5
r = x%2
if r==0:
    print("Even number")
else:
    print("Odd Number")
```

```python
print("Hello World\n"*8)
```

```python
i=0 
while i <5:
    print("Data Science -> ",i)
    i=i+1 # incremental while loop
```

```python
i = 5
while i >=1:
    print("Data Science -> ",i)
    i=i-1 # decremental while loop 
```

```python
for i in range(10):
    print(i)
```

```python
limit =5

for i in range(limit):
    print("index -> ",i)

```

```python
print("*"*50)
for i in range(limit,0,-1):
    print("Step -1 : reverse index -> ",i)
```

```python

print("*"*50)
for i in range(0,limit,2):
    print("Step 2 : index -> ",i)
```

```python
print("*"*50)
for i in range(0,limit,3):
    print("Step 3 : index -> ",i)
```
