```python
emp_df['Name'] = emp_df['Name'].str.replace(r'\W','',regex=True) # replace all the non character words
emp_df
```

```python
emp_df['Domain'] = emp_df['Domain'].str.replace(r"\W","",regex=True) # r'\W' : replace reqular expressions
emp_df
```

```python
emp_df['Location'] = emp_df['Location'].str.replace(r"\W","",regex=True)
emp_df
```

```python
emp_df['Age'] = emp_df['Age'].str.replace(r"\W","",regex=True)
emp_df
```

```python
emp_df['Age'] = emp_df['Age'].str.extract("(\d+)") # r(r'\d+')) : match digits only and ignore text/regx
emp_df
```
