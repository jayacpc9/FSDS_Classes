# Sorting set using sorted()

```python
name="Jaya Chandra Poreddy Channareddy"
nameSet=set(name) # taking all characters to lower, 
print(f"Name  = {name}")
print(f"All Unique characters are : {nameSet}")
sorted(nameSet) # note uppper case letters will preceed the lower case.
```

```python
name="JayaChandraPoreddyChannareddy"
nameSet=set(name.lower()) # taking all characters to lower, 
print(f"Name  = {name}")
print(f"All Unique characters are : {nameSet}")
sorted(nameSet) # note space and uppper case letters will preceed the lower case.
```

```python
a1 = {1, 2, 3, 4, 5}
b1 = {4, 5, 6, 7, 8}
c1 = {8, 9, 10}

print(f"A1 = {a1}")
print(f"B1= {b1}")
print(f"C1 = {c1}")
```

```python
a1.intersection_update(b1) # get the common elements from a1 and b1 and update it to a1, and remove others.

print(f"A1 = {a1}")
print(f"B1= {b1}")
print(f"C1 = {c1}")
```
