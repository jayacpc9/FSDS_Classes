# Break, Continue

```python
for i in range(1,11):
    if i == 6:
        break # break the loop when the condition is statisfied.
    print(i)
```

```python
for i in range(1,11):
    if i == 6:
        continue # continue the loop without the condition, so 6 is missing
    print(i)
```

# for else loop

```python
numbers = [2,4,6,9,10,12,15,17,18,20]
for num in numbers:
    if num % 8 == 0:
        print(num,end= "  ")
        continue
else:
    print("Number not found")

```

# check if the given number is prime or not

```python
num = 2

# Start at 2, stop before 12
for i in range(2, num):
    if (num % i == 0):
        print(f"{num} is not prime")
        break
else:
    # This else block runs ONLY if the loop finishes without hitting the 'break'
    print(f"{num} is prime")
```
