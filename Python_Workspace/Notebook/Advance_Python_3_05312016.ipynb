# check if the given number is prime or not

```python
num = 5

# Start at 2, stop before 12
for i in range(2, num):
    if (num % i == 0):
        print(f"{num} is not prime")
        break
else:
    # This else block runs ONLY if the loop finishes without hitting the 'break'
    print(f"{num} is prime")
```

```python
users = ["Alice", "Bob", "Charlie"]
target = "Bob"

for user in users:
    if user == target:
        print("User found! Logging in.")
        break
else:
    # This block handles the "Not Found" condition perfectly
    print("User not found, creating a new account.")
```
