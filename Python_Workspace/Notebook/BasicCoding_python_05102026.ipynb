```python
# in this case math is a module and has many inbuilt functions , for example square root(sqrt())
import math as m  # math is the module that is imported as m , m is an alias for math

x = m.sqrt(25)
print(x)
print(round(x))
```

```python
print(f"Floor value of 2.3 = {m.floor(2.3)}") # gets the nearest lower value
print(f"Floor value of 3.5 = {m.floor(3.5)}") # gets the nearest lower value
```

```python

print(f"Ceil value of 2.3 = {m.ceil(2.3)}") # gets the next highest value
print(f"Ceil value of 3.5 = {m.floor(3.5)}") # gets the next highest value
```

```python
from math import *
print(pow(2,3))
print(floor(2.5))
print(ceil(3.6))
print(sqrt(25))
print(round(sqrt(25)))
```

```python
x = int(input("Enter A : ")) # here the input is typecasted to int.
y = int(input ("Enter B : "))
print( f"{ x } + {y } = {x+y}") # since the input is of type int the addition happens correctly.as_integer_ratio
```
