```python
# python indexing
# Forward indexing
str = "Welcome"
print(str[0])
print(str[6])
print(str[4])
print(len(str))

# Backward indexing
print(str[-1])
print(str[-3])
print(str[-5])
print(str[-0])

# Error condition
# print(str[14]) # IndexError: string index out of range

```

```python
# String Slicing
s = "Welcome"
print(s[:])
print(s[2:4])
print(s[2:])
print(s[:4])
```

```python
# Advance string slicing
s ="Welcome"
print(len(s))
print(s[1:6]) # forward slicing
print(s[2:])
print(s[:7])
print(s[:len(s)//2]) # splits the string into half starting from index 0 to the middle
print(s[:100]) # no error cos it is slicing from 0 till 100, the length is 7 so it will just print all and leave it blank
```

```python
s = "Welcome"
print(s[-1:]) # Backward indexing
print(s[-7:])
print(s[-7: -4]) # Backward indexing
```

```python
# Data Types : numeric 
import sys
def printParamInt(val1): # function to print all the required vlaues of the given parameter
    print(val1)
    print(type(val1)) # type of object
    print(sys.getsizeof(val1)) # size of integer object in bytes
    print(val1, " is Integer?", isinstance(val1, int)) # val1 is an instance of int    

valueI = 10
printParamInt(valueI)
```
