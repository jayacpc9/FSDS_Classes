# How to print a file from python

```python
import subprocess

def print_file_unix(file_path):
    try:
        # Calls the system 'lp' command to send the file to the default printer
        subprocess.run(["lp", file_path], check=True)
        print("Job sent successfully!")
    except subprocess.CalledProcessError as e:
        print(f"Printing failed: {e}")
```

# How to print string content/data directly from python

```python
import subprocess

def print_string_unix(text_data):
    try:
        # Runs 'lp' and passes the string directly into its stdin
        subprocess.run(["lp"], 
                       input=text_data, 
                       text=True,# Tells Python to treat input as a string, not bytes
                       check=True
        )
        print("String sent to printer successfully!")
    except subprocess.CalledProcessError as e:
        print(f"Printing failed: {e}")

```

```python
# Example usage:
my_string = "Hello, this is a direct string print job from my Python code."
print_string_unix(my_string)
```
