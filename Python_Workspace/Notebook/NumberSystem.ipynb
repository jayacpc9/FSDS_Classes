# Number System : There are 4 number systems namely : 
### Decimal Number System: Base value is 10, asigned with 0x , valeu is 0 - 9
### Binary Number System : Base value is 2, assigned with 0b, valeu is 0 - 2
### Octal Number System: Base value is 8, assigned with 0o, valeu is 0 - 8
### Hexadecimal Number System: Base value is 16, assigned with 0x, valeu is 0 - 9, a-f


# Decimal Numbers

```python
num = 25
bin(num)
```

```python
int(0b11001)
```

# Binary Numbers : divide number by 2 and pick remainders from bottom up order.
### See example : to convert value dynamically
### int(Value, base)

```python
binVal = bin(30)
print(binVal)
originalVal = int(binVal,2)
print(originalVal)
```

# Octal Numbers : divide by 8

```python
num = 25
octVal=oct(num)
print(octVal)
originalVal = int(octVal,8)
print(originalVal)
```

```python
numb2 = 7
octVal=oct(numb2)
print(octVal)
originalVal = int(octVal,8)
print(originalVal)
```

# HexaDeecimal Numbers : divide by 16, value can be from 0 - 9, a - f
#### all ip addresses are written in hexa decimal numbers.

```python
num = 25
hexVal = hex(num)
print(hexVal)

originalVal = int(hexVal, 16)
print(originalVal)
```

# Swaping 2 Variables 
### For example :  if  (a,b) = (5,6)  swap them as (a,b) = (6,5)

```python
a = 5
b = 6
print(f"*"*10+" Before")
print(f"a = {a}")
print(f"b = {b}")

# use a temp variable 
temp = a
a =b
b = temp

print(f"*"*10+" After")
print(f"a = {a}")
print(f"b = {b}")
```

# Swapping without using a 3rd variable

```python
a = 5
b = 6
print(f"*"*10+" Before")
print(f"a = {a}")
print(f"b = {b}")

# use a temp variable 

a = a +b
b = a - b
a =  a - b

print(f"*"*10+" After")
print(f"a = {a}")
print(f"b = {b}")
```

# Swapping 2 numbers using XOR 

```python
a = 51232524525
b = 6986987692345

print(f"*"*10+" Before")
print(f"a = {a}")
print(f"b = {b}")

binA = bin(a)
binB = bin(b)
print(f"a = {binA}")
print(f"b = {binB}")


a = a ^ b
print(f" xor ( a = a ^ b ) = {a}")

b = a ^ b
print(f" xor ( b = a ^ b ) = {b}")

a = a ^ b
print(f" xor ( a = a ^ b ) = {a}")

# use a temp variable 
print(f"*"*10+" After")
print(f"a = {a}")
print(f"b = {b}")
```

# Bitwise Operators
## We have 6 Bitwise operators : 
### ~  - Compliment
### & - AND 
### | - OR 
### ^ - XOR 
### >> - Right Shift 
### << - Left Shift

```python
~12
```

# Bitwise AND Operator  : &

```python
a = 10
b = 12
print(bin(a))
print(bin(b))
print( a & b)
```

# Bitwise OR Operator  : |

```python
a = 12
b = 13
print(bin(a))
print(bin(b))
print( a | b)
```

# Bitwise XOR Operator  : ^

```python
a = 15
b = 14
print(bin(a))
print(bin(b))
print( a ^ b)
```

```python
a = 25
b = 30
print(bin(a))
print(bin(b))
print( a ^ b)
```

# Bitwise Left Shift Operator  : <<

```python
a = 13
b = 1
print(bin(a))
print(bin(b))

print( a<<b)
```
