# Functions

```python
# find the count of odd and even numbers form a given list
def count(lst):
    even=0
    odd = 0
    for i in lst:
        if(i%2==0):
            even+=1
        else:
            odd+=1
    return even,odd

lst = [10,9,8,23,50,34,100] 

even,odd = count(lst)

print("Even = ",even," Odd = ",odd)

print("Even Number count = {} and Odd Number count = {}".format(even,odd))
```

# Fibonacci Sequence

```python
def fibo(n):
    
    previous_fibo=0
    next_fibo=1
    fibo=0
    
    index=0
    print(previous_fibo)
    print(next_fibo)
    while index < n:
        fibo = previous_fibo+ next_fibo
        print(fibo)
        previous_fibo = next_fibo
        next_fibo =fibo
        index+=1

fibo(10)
```

```python
# factorial recursion
def factorial(n):
    if n==0:
        return 1
    else:
        return n * factorial(n-1)
        

n=5
result = factorial(n)

print("Factorial of {} is {}".format(n,result))
```

```python
nums = [3,2,6,8,4,6,2,9]

evens = list(filter(lambda n : n%2 ==0, nums))
double = list(map(lambda n : n+2, evens))

print(evens)
print(double)
```

```python
from functools import reduce

def add_all(a,b):
    return a+b

nums = [3,2,6,8,4,6,2,9]

evens = list(filter(lambda n : n%2 ==0, nums))
double = list(map(lambda n : n+2, evens))
sum = reduce(add_all,double)
print(evens)
print(double)
print(sum)
```

# Python Decorator

A decorator in Python is a function that takes another function as an argument, extends or modifies its behavior without changing the original function's code, and returns a new function.

Decorators are applied using the @decorator_name syntax placed directly above the function definition.

Why Use Decorators?

Decorators help you:

    Reuse code (DRY Principle)
    Add functionality without modifying existing code
    Keep business logic clean and separate
    Execute code before and/or after a function runs


# Basic Syntax


```python

def decorator(function):
    def wrapper():
        # Code before function call
        function()
        # Code after function call
    return wrapper

@decorator
def greet():
    print("Hello!")

greet()
```

How It Works

greet()
    │
    ▼
wrapper()
    │
    ├── Before Logic
    │
    ├── greet()
    │
    └── After Logic

The decorator wraps the original function, allowing additional code to run before and after it.

```python
#Example 1: Basic Decorator

def my_decorator(func):

    def wrapper():
        print("Before function execution")
        func()
        print("After function execution")

    return wrapper


@my_decorator
def say_hello():
    print("Hello!")


say_hello()
```

Example 2: Decorator with Arguments

Use *args and **kwargs to make the decorator work with any function.


```python
def log_decorator(func):

    def wrapper(*args, **kwargs):
        print(f"Calling {func.__name__}")

        result = func(*args, **kwargs)

        print(f"Result = {result}")

        return result

    return wrapper


@log_decorator
def add(a, b):
    return a + b


add(5, 10)

```
