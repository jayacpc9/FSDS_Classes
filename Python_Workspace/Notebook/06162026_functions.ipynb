# functions in python

```python
def greet():
    print("hello,good morning!!")

greet()
```

```python
# function to add any 2 numbers and return the sum
def add(x,y): # x and y are formal arguments
    c = x + y 
    print(f"The sum of {x} + {y} = {c}")

add(10,20) # 10,20 are called actual arguments.
add(50,100)
```

```python
def add(x,y): # x and y are formal arguments
    c = x + y 
    print(f"The sum of {x} + {y} = {c}")
    return c

r = add(10,20)
print(f"The return value r ={r}")
print(f"type(r)  = {type(r)}")
```

```python
def add(x,y,z): # x, y and z are formal arguments
    c = x + y + z 
    print(f"The sum of {x} + {y} + {z} = {c}")
    return c

r = add(3,4,5)
print(f"vale of r = {r}")
```

```python
def add(x,y,z=5): # x, y and z are formal arguments, with z having a default value and 
    c = x + y + z 
    print(f"The sum of {x} + {y} + {z} = {c}")
    return c

r = add(3,4)
print(f"vale of r = {r}")
```

```python
def add(x,y,z=5): # x, y and z are formal arguments, with z having a default value and 
    c = x + y + z 
    print(f"The sum of {x} + {y} + {z} = {c}")
    return c


def greet():
    return "hello,good morning!!"




r1 = add(3,4)
r2 = greet()
print(f"vale of r1 = {r1}")
print(r2)

print(f"type(r1) = {type(r1)}")
print(f"type(r2) = {type(r2)}")
```

```python
def add(x,y):
    c = x + y 
    d = x - y 
    return c,d


def greet():
    return "hello,good morning!!"


r1 = add(20,5) # r1 is a tuple contains 2 results

print(f"len(r1) = {len(r1)}")
print(f"r1[0] = {r1[0]} and r1[1] = {r1[1]}")
print(f"type(r1) = {type(r1)}")


r1,r3 = add(20,5) #accept the tuple in 2 variables

print(f"r1 = {r1} and r3 = {r3}")
r2 = greet()
print(r2)

print(f"type(r1) = {type(r1)}")
print(f"type(r2) = {type(r2)}")
```

# actual arguments is dividedn into 4 parts 
#### 1. position 
#### 2. keyword 
#### 3. default
#### 4. variable length arguments

# 1. positional Arguments

```python
def person(name, age):
    print("name = ",name)
    print("age = ",age)

person("joe",45)
print("-"*20)
person(45,"joe")
```

# 2.Keyword arguments

```python
def person(name, age):
    print("name = ",name)
    print("age = ",age)

person(age=45,name="joe")
```

# you can use either a position argument or keyword argument method,

```python
def person(name, age,mobile):
    print("name = ",name)
    print("age = ",age)

person(age=45,name="joe",mobile=1234)
```
