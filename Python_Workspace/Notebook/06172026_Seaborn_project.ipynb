```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,6))
```

```python
# scatterplot
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="time", size="size", palette="deep")
plt.title("Tips vs total bill")
plt.show()
```

```python
# lineplot 
sns.lineplot(data=tips, x="size", y="total_bill", hue="sex", markers="o")
plt.title("Lineplot of Total Bill vs Size")
plt.show()
```

```python
# lineplot 
sns.lineplot(data=tips, x="size", y="total_bill", hue="sex", markers="o", ci=None)
plt.title("Lineplot of Total Bill vs Size")
plt.show()
```

```python
tips.columns
```

```python
#  barplot
sns.barplot(data=tips,x="day", y="total_bill", hue= "sex", palette="muted")
plt.title("Barplot of Total Bill by day")
plt.show()
```

```python
sns.boxplot(data=tips,x="day", y="tip", hue="smoker", palette="Set2")
plt.title("Boxplot of Tips by Day and Smoker Status")
plt.show()
```

```python
# violin plot
sns.violinplot(data=tips,x='day', y='total_bill', hue='time', split=True, palette="pastel")
plt.title("Violin Plot of Total Bill by Day and Time")  
plt.show()
```

```python
# countplot
sns.countplot(data=tips, x='day', hue='smoker', palette='dark')
plt.title("Count Plot of Days by Smoker Status")
plt.show()
```

```python
tips.columns
```

```python
sns.regplot(data=tips, x='total_bill', y='tip', scatter_kws={'s':50}, line_kws={'color':'red'})
plt.title("Count Plot of Days by Smoker Status")
plt.show()
```

```python
sns.histplot(data=tips, x='total_bill', bins=20, kde=True, color='blue')
plt.title("Histogram of Total Bill with KDE")
plt.show()
```

```python
# pairplot
sns.pairplot(data=tips, hue='sex', vars=['total_bill','tip','size'],palette='bright')
plt.suptitle("pair plot: numeric variables by gender", y=1.02)
plt.show()
```

```python
# catplot
sns.catplot(data=tips,x='day',y='tip', hue='sex',kind='point', palette='dark')
plt.title("catplot(point):Tips by day and gender")
plt.show()
```

```python
#jointplot
sns.jointplot(data=tips,x='total_bill',y='tip', kind='scatter', hue='smoker',color='purple', palette='bright')
# sns.jointplot(data=tips, x='total_bill', y='tip', kind='scatter', hue='smoker',  palette='coolwarm') # remove color
plt.suptitle("Jointplot: Total Bill vs Tip",y=1.05)
plt.show()
```

```python
# FacetGrid
sns.FacetGrid(tips, col='time', row='smoker', margin_titles=True).map(sns.histplot, 'total_bill', bins=20, kde=True).add_legend()

```

```python
sns.stripplot(data=tips, x='day',y='tip', hue='sex', jitter=True, palette='muted')
plt.title("strip plot: Tips by data and gender")
plt.show()
```

```python

sns.kdeplot(data=tips, x='total_bill', y='tip', hue='sex')
```

```python
sns.kdeplot(data=tips, x='total_bill', y='tip', hue='sex', fill=True, palette='tab10')
```

```python
sns.kdeplot(data=tips, x='total_bill',hue='sex', fill=True, palette='tab10')
plt.title("kde plot:Total bill density by gender")
plt.show()
```
