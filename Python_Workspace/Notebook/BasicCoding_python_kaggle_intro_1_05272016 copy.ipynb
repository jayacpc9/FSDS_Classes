# Work around for Pagination
# logic to iterate between index based on page number and items per page
# 

```python
# iterate through all the movie detials
# action_movie_details data frame will contain only the 
# 
page_id = 3
no_of_items_per_page = 3
    
start_index = (page_id - 1) * no_of_items_per_page # always use the previous page (page_id -1)
end_index = page_id * no_of_items_per_page


for index, movie_details in action_movie_details.iloc[start_index:end_index].iterrows():
    print(f"Index:\t\t{index}")
    print(f"Movie ID:\t{movie_details['movieId']}")
    print(f"Title:\t\t{movie_details['title']}")
    print(f"Genres:\t\t{movie_details['genres']}")
    print("-" * 40)
```
