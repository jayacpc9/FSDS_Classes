In this practice problem, we want you to extract all even values from a 2-dimensional array. The result should be a 1-dimensional array that contains all even values of the provided 2-dimensional array.

Recall that the even numbers are 0, 2, 4, 6, and so on.

For example, if the 2-dimensional array was:

[[1 2 4]

 [3 4 6]]

```python
# Answer
all_values=np.array([[1,2,4],[3,4,6]])
even_values = all_values[all_values % 2 ==0]
mask = [all_values % 2 ==0]
print(f"Mask = {mask}")
# even_values = all_values[mask]

print(f"The original matrix = \n{all_values}\n")
print(f"The resultant even numbers = {even_values}")

```

```python
import numpy as np

# Initial setup
arr = np.array([[1, 2, 4], 
                [3, 4, 6]])

print("--- Step 0: Original Array ---")
print(arr)

# STEP 1: Create the Boolean Mask
# This evaluates the condition for every element
mask = (arr % 2 != 0)
print("\n--- Step 1: The Mask (Identifying Odds) ---")
print(mask)

# STEP 2: Use the mask to select elements
# We can see what the mask "sees" before we change it
odds_detected = arr[mask]
print("\n--- Step 2: Elements identified by Mask ---")
print(odds_detected)

# STEP 3: Assignment (In-place modification)
# This replaces only the 'True' positions with 0
arr[mask] = 0
print("\n--- Step 3: Result after Assignment ---")
print(arr)
```
