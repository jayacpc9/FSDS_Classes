# NumPy : Advance functionalities

```python
import numpy as np
```

```python
# create an array from list
list = [1,2,3]
a = np.array(list)
print("Array a = ",a)
```

```python
# create an array with evenly spaced balues
b= np.arange(0,10,3) # values from 0 to 10 with step 3
print("Array b = ",b)
```

```python
# create an array filled with zeros
d = np.zeros((2,3)) # create a 2X3 array of zeros as default values, adds floats by default
print("Arrau d: \n",d)

# this is example for parameter
```

```python
# create an array filled with zeros
d = np.zeros((2,3), dtype=int) # create a 2X3 array of zeros as default values, dtype as int for whole number
print("Array d: \n",d) 

# this is example for hyperparameter
```

```python
e = np.ones((3,2)) # 3X2 array of ones as default values, with default float values
print("Array e = ",e)
```

```python
e = np.ones((3,2), dtype=int) # 3X2 array of ones as default values, with default int values
print("Array e = ",e)
```

```python
# create an identity matrix
f = np.eye(4) # a matrix whose diagnol elements are all 1, is called identity matrix
print("Identity matrix = ",f)
```

# Array Manipulation Functions

```python
# Reshape an array
a1= np.array([1,2,3])
print("Original Array = ",a1)
print("Original Array shape = ",a1.shape)
reshaped = np.reshape(a1,(1,3))
print("Reshaped array = ",reshaped)
print("Reshaped array shape = ",reshaped.shape)
```

```python
# Reshape an array
a1= np.array([1,2,3])
print("Original Array = ",a1)
print("Original Array shape = ",a1.shape)
reshaped = np.reshape(a1,(3,1))
print("Reshaped array = \n",reshaped)
print("Reshaped array shape = ",reshaped.shape)
```

```python
# Flatten an array
f1 = np.array([[1,2],[3,4],[5,6],[7,8],[8,9],[9,10]])
print("f1 array = \n",f1)
flattened = np.ravel(f1)
print("Flattened array = ",flattened)
```

```python
# Transpose an array
e1 = np.array([[1,2],[3,4]])
transposed = np.transpose(e1)
print("Transposed array = \n",transposed)
```
