# Array Manipulation Functions

```python
# Reshape an array
a1= np.array([1,2,3])
print("Original Array = ",a1)
print("Original Array shape = ",a1.shape)
reshaped = np.reshape(a1,(1,3))
print("Reshaped array = ",reshaped)
print("Reshaped array shape = ",reshaped.shape)
```

```python
# Reshape an array
a1= np.array([1,2,3])
print("Original Array = ",a1)
print("Original Array shape = ",a1.shape)
reshaped = np.reshape(a1,(3,1))
print("Reshaped array = \n",reshaped)
print("Reshaped array shape = ",reshaped.shape)
```

```python
# Flatten an array
f1 = np.array([[1,2],[3,4],[5,6],[7,8],[8,9],[9,10]])
print("f1 array = \n",f1)
flattened = np.ravel(f1)
print("Flattened array = ",flattened)
```

```python
# Transpose an array
e1 = np.array([[1,2],[3,4]])
transposed = np.transpose(e1)
print("Transposed array = \n",transposed)
```

```python
# stack arrays vertically
a2 = np.array([1,2])
b2 = np.array([3,4])
stacked = np.vstack([a2,b2])
print("Vertical stacking = \n",stacked)
```

```python
# stack arrays Horizontally
a2 = np.array([1,2,])
b2 = np.array([3,4])
stacked = np.hstack([a2,b2])
print("Horizontal stacking = \n",stacked)
```

# Statistical Functions

```python
s = np.array([1,2,3,4])
mean = np.mean(s)
print(f"The mean value of  s  = {mean}")
```

```python
# standard deviation of an array
s = np.array([1,2,3,4,5])
var = np.var(s)
print("The values = ",s)
print("Variance of s = ", var)
```

```python
# standard devication of an array
print("The values = ",s)
std_dev = np.std(s)
print("Standard deviation of s: ",std_dev)
```

```python
# minimum element of an array
minimum = np.min(s)
print("The minimum value = ", minimum)
```

```python
# Maximum element of an array
maximum = np.max(s)
print("The maximumm value = ", maximum)
```

# Set Operation

```python
# Intersection of two arrays
set_a = np.array([1,2,3,4])
set_b= np.array([3,4,5,6])
intersection = np.intersect1d(set_a, set_b)
print("Ther intersect1d of set_a and set_b = ",intersection)
```

```python
set_a = np.array([1,2,3,4])
set_b= np.array([3,4,5,6])
union_val = np.union1d(set_a, set_b)
print("Ther union1d of set_a and set_b = ",union_val)
```
