# Print 

```python
a = 10
b = 20
print(a)
print(b)
```

```python
print("The value of A = ",a)
print("The value of B = ",b)
```

```python
print("The sum of ",a," + ",b, " = ",(a+b))
```

# printing using format method

```python
print("the sum of {} and {} = {}".format(a,b,(a+b)))
```

# printing using f String

```python
print(f"The sum of {a} + {b} = {a+b}")
```

# end statements

```python
print("Hello team", end = ' ') # start generating new line
print("Good Morning")
```

```python
print("Hello team", end = ' ## ') # start generating new line
print("Good Morning")
```
